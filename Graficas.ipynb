```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Datos de la tabla
datos = {
    "Instancia": ["MDG-a_1_n500_m50.txt", "MDG-b_21_n2000_m200.txt", "MDG-c_1_n3000_m300.txt"],
    "GA-B (s)": [17.1, 260.1, 695.9],
    "GA-O (s)": [5.1, 134.3, 280.5],
    "MA (s)": [5.5, 166.5, 433.6]
}

df = pd.DataFrame(datos)

# Cálculo del porcentaje de reducción respecto a GA-B
df["Reducción GA-O (%)"] = ((df["GA-B (s)"] - df["GA-O (s)"]) / df["GA-B (s)"]) * 100
df["Reducción MA (%)"] = ((df["GA-B (s)"] - df["MA (s)"]) / df["GA-B (s)"]) * 100

# Mostrar los datos actualizados
print(df)
```

```python
# Datos
instancias = df["Instancia"]
tiempo_GA_B = df["GA-B (s)"]
tiempo_GA_O = df["GA-O (s)"]
tiempo_MA = df["MA (s)"]
reduccion_GA_O = df["Reducción GA-O (%)"]
reduccion_MA = df["Reducción MA (%)"]

# Crear gráfico de barras comparando tiempos de ejecución
fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.3
index = np.arange(len(instancias))

plt.bar(index, tiempo_GA_B, bar_width, label="GA Base", color="blue")
plt.bar(index + bar_width, tiempo_GA_O, bar_width, label="GA Optimizado", color="green")
plt.bar(index + 2 * bar_width, tiempo_MA, bar_width, label="Memético", color="red")

plt.xlabel("Instancia")
plt.ylabel("Tiempo de Ejecución (s)")
plt.title("Comparación de Tiempos de Ejecución")
plt.xticks(index + bar_width, instancias, rotation=15)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar gráfico
plt.show()
```

```python
# Crear gráfico de reducción porcentual
fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.3

plt.bar(index, reduccion_GA_O, bar_width, label="Reducción GA-O", color="green")
plt.bar(index + bar_width, reduccion_MA, bar_width, label="Reducción MA", color="red")

plt.xlabel("Instancia")
plt.ylabel("Reducción de Tiempo (%)")
plt.title("Reducción de Tiempo con Respecto a GA Base")
plt.xticks(index + bar_width / 2, instancias, rotation=15)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar gráfico
plt.show()
```

## Paralelización

```python
# Datos de la tabla
datos = {
    "Instancia": ["MDG-a", "MDG-b", "MDG-c"],
    "MA sin paralelizar": [3600, 13760, 6330],
    "MA paralelizado": [264.3, 2540.4, 2606.6]
}

df = pd.DataFrame(datos)

# Cálculo del porcentaje de reducción respecto a GA-B
df["Reducción MA (%)"] = ((df["MA sin paralelizar"] - df["MA paralelizado"]) / df["MA sin paralelizar"]) * 100

# Mostrar los datos actualizados
print(df)
```

```python
# Datos
instancias = df["Instancia"]
tiempo_MA_no_paralelizado = df["MA sin paralelizar"]
tiempo_MA_paralelizado = df["MA paralelizado"]
reduccion_MA = df["Reducción MA (%)"]

# Crear gráfico de barras comparando tiempos de ejecución
fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.3
index = np.arange(len(instancias))

plt.bar(index, tiempo_MA_no_paralelizado, bar_width, label="Sin paralelizar", color="blue")
plt.bar(index + bar_width, tiempo_MA_paralelizado, bar_width, label="Paralelizado", color="red")

plt.xlabel("Instancia")
plt.ylabel("Tiempo de Ejecución (s)")
plt.title("Comparación de Tiempos de Ejecución")
plt.xticks(index + bar_width, instancias, rotation=15)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar gráfico
plt.show()
```

```python
# Crear gráfico de reducción porcentual
fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.3

plt.bar(index, reduccion_MA, bar_width, label="Reducción paralelizacion", color="green")

plt.xlabel("Instancia")
plt.ylabel("Reducción de Tiempo (%)")
plt.title("Reducción de Tiempo con Respecto a GA Base")
plt.xticks(index + bar_width / 2, instancias, rotation=15)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Mostrar gráfico
plt.show()
```
